# **Task 2:**
You must have noticed we have some data with missing "Morningstar Ratings", createa method and figure out a way to imputate the missing data

```python
mean_MSR = data['Morningstar Rating'].mean()
print(mean_MSR)
def remove_nan_MSR(row):
  if pd.isnull(row ['Morningstar Rating']):
    return mean_MSR
  else:
    return row['Morningstar Rating']
  
```

```python
data['Morningstar Rating']= data.apply(remove_nan_MSR, axis=1 )
```

# **Task 4:**
Mapping your data: Figure out a way to map the values of "Morningstar Rating" form [5,4,3,2,1] to ['Awsome','Good','Average','bad','Awful']

```python
#to remove the decimal points we used the round off method so that we can categorise the data according to question.

data['Morningstar Rating'] = pd.Series([round(val,0) for val in data['Morningstar Rating']], index = data.index)
```

```python
data['Morningstar Rating_Name'] = data['Morningstar Rating'].map(lambda s :  'Awesome' if s==5.0 else 'Good' if s==4.0 else 'Average' if s==3.0 else  'bad' if s==2.0 else 'Awful')
data[['Morningstar Rating_Name' , 'Morningstar Rating']]
```

```python
data['Morningstar Rating'] = pd.Series([round(val,3) for val in data['Morningstar Rating']], index = data.index)

data["Morningstar Rating"]
```
